```python
perf = soup.findAll(attrs = {'class':'tribe-event-url'})
print(type(perf))
perf
```

##### attrs allows you to be more specific with your findAll. 

```python
headliner = [x.get('title') for x in perf] ##string method functions as shorthand for loop
headliner
```

```python
for y in perf:
    test = y.get('title')
    print(test)
```

```python
headlinerdf = pd.DataFrame(headliner, columns = ['artist'])
headlinerdf.head(1)
```

```python
page = 1
perfbig = []
timebig = []
datebig = []

while page != 6:
    url = f"https://ryman.com/events/list/?tribe_event_display=list&tribe_paged={page}"
    response = requests.get(URL)
    soup = BS(response.text)
    
    for y in soup.find_all(attrs = {'class':'tribe-event-url'}):
        perfbig.append(y.get('title'))
        
    for x in soup.find_all('time'):
        x = x.get('datetime')
        x = x.split(' ', 1)
        datebig.append(x[0])
        timebig.append(x[1])
    page = page + 1
    
print(perfbig) 
print(timebig)
print(datebig)
```
